# Let's create a  Graphical User Interface (GUI) that will allow users input the four independent variables in order to get the predicted result(KPI). 

## We'll be using the Tkinter package, the standard Python interface to the Tk GUI toolkit.

```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
# tkinter GUI
root= tk.Tk()
```

```python
root= tk.Tk()

canvas1 = tk.Canvas(root, width = 1200, height = 450)
canvas1.pack()

# with sklearn
Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
canvas1.create_window(260, 220, window=label_Intercept)

# with sklearn
Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
canvas1.create_window(260, 240, window=label_Coefficients)

# with statsmodels
print_model = model.summary()
label_model = tk.Label(root, text=print_model, justify = 'center', relief = 'solid', bg='LightSkyBlue1')
canvas1.create_window(800, 220, window=label_model)

'GDN_Spend','Google_nonbrand_Spend','Google_nonbrand_IMP','Google_brand_imp'

# New_GDN_Spend label and input box
label1 = tk.Label(root, text='Type GDN Spend: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)

# New_Google_nonbrand_Spend label and input box
label2 = tk.Label(root, text=' Type GNB Spend: ')
canvas1.create_window(100, 120, window=label2)

entry2 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 120, window=entry2)

# New_Google_nonbrand_IMP label and input box
label3 = tk.Label(root, text='Type GNB IMP: ')
canvas1.create_window(100, 140, window=label3)

entry3 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 140, window=entry3)

# New_Google_brand_imp label and input box
label4 = tk.Label(root, text='Type GB IMP: ')
canvas1.create_window(100, 160, window=label4)

entry4 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 160, window=entry4)

def values(): 
    global New_GDN_Spend #our 1st input variable
    New_GDN_Spend = float(entry1.get())
    
    global New_Google_nonbrand_Spend #our 2nd input variable
    New_Google_nonbrand_Spend = float(entry2.get())
    
    global Google_nonbrand_IMP #our 3rd input variable
    Google_nonbrand_IMP = float(entry3.get())
    
    global Google_brand_imp #our 4th input variable
    Google_brand_imp = float(entry4.get())
    
    Prediction_result  = ('Predicted KPI: ', regr.predict([[New_GDN_Spend ,New_Google_nonbrand_Spend,Google_nonbrand_IMP,Google_brand_imp]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 280, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict KPI',command=values, bg='orange') # button to call the 'values' command above 
canvas1.create_window(270, 190, window=button1)

root.mainloop()
```
